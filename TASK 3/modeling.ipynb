Everything looks fine for the 3 datasets, except that, if we want to extract information about time, the timestamp should be a datetime object.

```python
def convert_to_datetime(df: pd.DataFrame, column: str):
    
    dummy = df.copy()
    dummy[column] = pd.to_datetime(dummy[column], format='%Y-%m-%d %H:%M:%S')
    return dummy
```

The function is now used to obtain the datasets in the desired format:

```python
sales_df = convert_to_datetime(sales_df, 'timestamp')
stock_df = convert_to_datetime(stock_df, 'timestamp')
temp_df = convert_to_datetime(temp_df, 'timestamp')

print(sales_df['timestamp'].dtype)
```

## Section 4 - Dataframe merging

Revisiting the clients needs, we need a model to predict the stock levels of products on an hourly basis. Therefore, to make things easier, the minute and second details of the timestamps are going to be set to 00:00.

```python
from datetime import datetime

def datetime_hourly(df: pd.DataFrame, column : str):
    dummy = df.copy()
    dummy[column] = dummy[column].dt.floor('H')

    return dummy


sales_df = datetime_hourly(sales_df,'timestamp')
sales_df.head()
stock_df = datetime_hourly(stock_df, 'timestamp')
stock_df.head()
temp_df = datetime_hourly(temp_df, 'timestamp')
temp_df.head()
```

The next step is to group the sales of the same products by the hour they were sold. Now we create groups where rows with the same combination of 'timestamp' and 'product_id' values are grouped together. Then, we sum the 'quantity' column of every column per group and obtain the quantity of a specific product that was sold at a specific hour.

```python
sales_agg = sales_df.groupby(['timestamp', 'product_id']).agg({'quantity': 'sum'}).reset_index()
sales_agg
```

The stock information is going to be also groupped by timestamp and product id, and aggregated by the mean estimated stock pct.

```python
stock_agg = stock_df.groupby(['timestamp', 'product_id']).agg({'estimated_stock_pct': 'mean'}).reset_index()
stock_agg
```

The storage temperature data is going to be grouped by timestamp and aggregated by temperature


```python
temp_agg = temp_df.groupby(['timestamp']).agg({'temperature': 'mean'}).reset_index()
temp_agg.head()
```

After that, the datasets will be merged by their common columns:
- Sales and stock can be merged by timestamp and product id
- Temperature can be merged to the other two by timestamp

Note that the predominant dataset is the stock_agg, as there might be some products that have not been yet sold and they are storaged. 

```python
merged_df = sales_agg.merge(stock_agg, on= ['timestamp', 'product_id'], how= 'right')
merged_df.head()
merged_df.info()
```

```python
merged_df = merged_df.merge(temp_agg, on= ['timestamp'], how= 'left')
merged_df.head()
```

```python
merged_df.info()
```

The column quantity lacks some values due to the reason that some products might not have been purchased at specific hours in the day. Therefore, those values should be filled up with 0s.

```python
merged_df['quantity'] = merged_df['quantity'].fillna(0)
merged_df.info()
```

Now, we can add more features to the dataset, such as the price of each product and its category. The sales data can contain repeated products that have been sold repeteadly. Hence, we will drop the duplicates. 


```python
product_categories = sales_df[['product_id', 'category']]
product_categories = product_categories.drop_duplicates() 

product_price = sales_df[['product_id', 'unit_price']]
product_price = product_price.drop_duplicates()
```

```python
merged_df = merged_df.merge(product_categories, on= 'product_id', how= 'left')
merged_df = merged_df.merge(product_price, on= 'product_id', how= 'left')

merged_df.head()
```

```python
merged_df.info()
```

Now we have a table with much more information about products and sales.

## Section 5 - Feature Engineering

This process is necessary to make our tables suitable for a machine learning model. We would need numeric columns for this case. The timestamp, for example, will be expanded into different columns.

```python
merged_df['day_of_month'] = merged_df['timestamp'].dt.day 
merged_df['day_of_week'] = merged_df['timestamp'].dt.day_of_week
merged_df['month'] = merged_df['timestamp'].dt.month
merged_df['hour'] = merged_df['timestamp'].dt.hour
merged_df.drop(columns= ['timestamp'], inplace= True)
merged_df.head()
```

We will also create dummy columns for the categories, as they all go by names, it is a way to number them:

```python
merged_df = pd.get_dummies(merged_df, columns=['category'])
merged_df.head()
```

```python
merged_df.info()
```

Having done this, it is notable now that the product_id column is cateogric and will not contribute to the success of our machine learning model, hence we remove it. 


```python
merged_df.drop(columns= ['product_id'], inplace= True)
merged_df.head()
```

Now we have a dataframe that only contains numeric values. We can then start building our model.

## Section 6 - Model Building

Now it is time to train our machine learning model with the data. For this task, we will use supervised learning. We want to be able to predict our estimated stock percentage variable. Thus, we create a target variable y that will contain the data for this variable, and a daframe x that will contain the merged_df, but without the estimated_stock_pct column.

```python
x = merged_df.drop(columns=['estimated_stock_pct'])
y = merged_df['estimated_stock_pct']

print(x.shape)
print(y.shape)
```

These results indicate that we have 29 different variables to make a prediction on our model. We will use sklearn's Random Forest Regressor. For that, we can create a class that contains methods to build the model, stablish hyperparameters, split the dataset, and evaluate the model.


```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler

class RegressionModel():
    
    def __init__(self, test_size = 0.25) -> None:
        self.model = RandomForestRegressor()
        self.test_size = test_size
        self.hyperparameters = None
        pass

    def train_test_split(self, X,y, test_size):

        ''' Implementing train-test split logic '''
        
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state= 0)

        scaler = StandardScaler().fit(X_train)
        X_train = scaler.transform(X_train)
        X_test = scaler.transform(X_test)
        train_set = (X_train, y_train)
        test_set = (X_test, y_test)

        
        return train_set, test_set
    
    def preprocess(self, X):

        '''Scaling data'''
        scaler = StandardScaler().fit(X)
        X = scaler.transform(X)
        return X
    
    def fit(self, train_set): 

        ''' Performs training logic for any model. First pre-processing, 
        and then training. '''
        
        X_train, y_train = train_set
        X_train = self.preprocess(X_train)     
        self.model.fit(X_train, y_train)

        return self.model
    
    def evaluate(self, test_set):

        '''Performs prediction of the model and stores metrics'''
        
        X_test, y_test = test_set
        X_test = self.preprocess(X_test)
        y_pred = self.model.predict(X_test)

        mae = mean_absolute_error(y_pred= y_pred, y_true= y_test)
        r2 = r2_score(y_pred= y_pred, y_true= y_test)

        return mae, r2
    
```

Now, having created that model, we use K-fold cross-validation. K-fold cross-validation is a crucial technique in machine learning, ensuring the robustness and reliability of models. By repeatedly partitioning the data into training and testing sets, it provides a more realistic assessment of a model's performance compared to a single train-test split. This approach minimizes the risk of overfitting and underfitting, maximizes data utilization, and aids in hyperparameter tuning and model selection

```python
accuracy = []
K = 10

for fold in range(0, K):

    model = RegressionModel()
    # Create training and test samples
    train_set, test_set = model.train_test_split(x,y, test_size= 0.25)
    model.fit(train_set)
    mae, r2 = model.evaluate(test_set)
    accuracy.append(mae)

    print(f"Fold {fold + 1}: MAE = {mae:.3f}")

print(f"Average MAE: {(sum(accuracy) / len(accuracy)):.2f}")
    
```

The fact that the MAE is almost equal every time suggests that the performance of the model is consistent across the different folds created, which is a good sign. However, we would need our MAE to be reduced ideally. That is why we should suggest the client to send us more datasets. We could also try to find ways to feature engineer better, but more data means more accuracy. 

We can now use the trained model to interpret what features are most important when the model was predicting the target variable. 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python


features = [i.split("__")[0] for i in x.columns]
importances = model.model.feature_importances_
indices = np.argsort(importances)

fig, ax = plt.subplots(figsize=(10, 20))
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

This feature importance visualisation tells us:

- The product categories were not that important
- The unit price and temperature were important in predicting stock
- The hour of day was also important for predicting stock
